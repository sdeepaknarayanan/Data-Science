```python
from sklearn.datasets import fetch_mldata
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
import keras
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
km = KMeans(n_clusters=10,init='random',max_iter=1000)
```

```python
train = []
for i in range(len(X_train)):
    temp = np.reshape(X_train[i],784)
    train.append(temp)
train = np.array(train)
```

```python
km.fit(train)
```

```python
test = []
for i in range(len(X_test)):
    test.append(np.reshape(X_test[i],784))
```

```python
labels = {i:[] for i in range(10)}
for i in range(len(train)):
    min_ = [100,np.inf]
    for j in range(len(km.cluster_centers_)):
        if np.linalg.norm(km.cluster_centers_[j] - train[i])<min_[1]:
            min_[1] = np.linalg.norm(km.cluster_centers_[j] - train[i])
            min_[0] = j
    labels[min_[0]].append(y_train[i])
```

```python
for i in labels:
    plt.figure()
    plt.hist(labels[i])
```

```python
new_labels = {}
for i in labels:
    data = Counter(labels[i])
    data.most_common()   # Returns all unique items and their counts
    print(i,data.most_common(1)[0][0]) # Returns the highest occurring item 
    new_labels[i] = data.most_common(1)[0][0]
```

```python
new_labels
```

```python
predicted = km.predict(test)
```

```python
accurate = 0
for i in range(len(predicted)):
    if new_labels[predicted[i]]==y_test[i]:
        accurate+=1
print("Accuracy is ",accurate/len(y_test)*100)
```
