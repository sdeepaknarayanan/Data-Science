### Implementation of the Push, and the Approximate PageRank Algorithms


```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import operator
```

```python
G = nx.Graph()
dataset = open('com-amazon.ungraph.txt','r')
for i in dataset:
    b = i.split()
    if b[0]!='#':
        G.add_edge(int(b[0]),int(b[1]))
G.remove_edges_from(nx.selfloop_edges(G))
dataset.close()
```

```python
l_nodes = list(G.nodes)
new_node = {}
for i in range(len(l_nodes)):
    new_node[l_nodes[i]] = i
```

```python
alpha = 0.2 
epsilon = 0.01
```

```python
def push(G,p, r, u):
    p_ = np.copy(p)
    r_ = np.copy(r)
    p_[new_node[u]] = p[new_node[u]] + alpha*r[new_node[u]]
    r_[new_node[u]] = (1-alpha)*r[new_node[u]]/2
    for nbr in G.neighbors(u):
        r_[new_node[nbr]] = r[new_node[nbr]] + (1-alpha)*r[new_node[u]]/(2*G.degree(u))
    return p_, r_

def condition(G,r,epsilon):
    nodes = list(G.nodes)
    flag = 0
    list_nodes = []
    for i in range(len(nodes)):
        if r[new_node[nodes[i]]]/G.degree(nodes[i])>=epsilon:
            list_nodes.append(nodes[i])
            flag = 1
    if flag==1:
        return 1,list_nodes
    else:
        return 0, []

def approximate_page_rank(G, v, alpha, epsilon):
    p = np.zeros(G.number_of_nodes())
    r = np.zeros(G.number_of_nodes())
    r[v] = 1
    is_sat, nodes = condition(G,r,epsilon)
    while(condition(G,r,epsilon)[0]):
        nodes = condition(G,r,epsilon)[1]
        u = np.random.choice(nodes)
        p,r  = push(G,p,r,u)
    return p
```

### Implementation of the Conducatance plots 

```python
seed_nodes = [514553 ,
4965 ,
143608,
243284,
152623 ,
533768,
59217 ,
99373 ,
89325 ]
new_list = [new_node[i] for i in seed_nodes]
```

```python
p_all = []
sup = {}
for i in range(5):
    p = approximate_page_rank(G,new_list[i],0.2,0.005)
    p_all.append(p)
    support = 0
    for j in range(len(p)):
        if p[j]!=0:
            support+=1
            if new_list[i] in sup:
                sup[new_list[i]].append([j])
            else:
                sup[new_list[i]]=[j]
    print(seed_nodes[i],"Support is",support)
```

```python
for i in sup:
    new = []
    for j in sup[i]:
        if type(j)==list:
            new.append(j[0])
        else:
            new.append(j)
    sup[i] = new
```

```python
sup_keys = list(sup.keys())
```

```python
node_conductance = {}
for i in new_node:
    for j in sup_keys:
        if new_node[i]==j:
            node_conductance[i]=[]
```

```python
for i in node_conductance:
    for j in sup[new_node[i]]:
        for k in new_node:
            if new_node[k]==j:
                node_conductance[i].append(k)
```

```python
p_mapping = {}
p_mapping[514553] = p_all[0]
p_mapping[4965] = p_all[1]
p_mapping[143608] = p_all[2]
p_mapping[243284] = p_all[3]
p_mapping[152623] = p_all[4]
```

```python
final = []
new_dict = {i:0 for i in node_conductance}

for i in node_conductance:
    new_list_ = []
    for j in range(len(node_conductance[i])):
        new_list_.append((node_conductance[i][j],p_mapping[i][new_node[node_conductance[i][j]]]/G.degree(node_conductance[i][j])))
    final.append(new_list_)
```

```python
for elem in range(len(final)):
    final[elem] = (sorted(final[elem],key=lambda x:(-x[1],x[0])))
```

```python
new_arr_cond = []
for i in range(len(final)):
    new_arr_cond.append([])
    for j in range(len(final[i])):
        new_arr_cond[i].append(final[i][j][0])
```

```python
conductance_cuts = {i:[] for i in range(len(final))}
for i in range(len(final)):
    for j in range(len(new_arr_cond[i])):
        try:
            temp = nx.algorithms.cuts.conductance(G,new_arr_cond[i][:j])
            conductance_cuts[i].append(temp)
        except:
            continue
```

```python
for i in conductance_cuts:
    plt.figure()
    len1 = len(conductance_cuts[i])
    plt.title('Conductance vs k for seed node '+str(i+1))
    plt.xlabel('k')
    plt.ylabel('Conductance')
    plt.plot(range(len1),conductance_cuts[i])
```

### Precision plots, obtained across multiple communities for each seed node - interpretation - just fraction of nodes in communities

```python
new_arr = {}
for i in range(len(final)):
    temp = []
    for j in range(len(final[i])):
        temp.append(final[i][j][0])
    new_arr[final[i][0][0]] = (temp)
```

```python
file_precision = open('com-amazon.all.dedup.cmty.txt')
presence = {i:[] for i in node_conductance}
for i in file_precision:
    arr = i.split()
    for i in range(len(arr)):
        arr[i] = int(arr[i])
    for j in presence:
        if j in arr:
            presence[j].append(arr)
```

```python
present = []
for i in node_conductance:
    for j in range(len(new_arr[i])):
        for k in range(len(presence[i])):
            if new_arr[i][j] in presence[i][k]:
                present.append((i,k,j))
```

```python
num_comm = {i:[] for i in node_conductance}
for i in new_arr:
    for j in present:
        if i==j[0]:
            if j[1] not in num_comm[i]:
                num_comm[i].append(j[1])
```

```python
new_num_comm = {i:{j:[] for j in num_comm[i]} for i in num_comm}
```

```python
for i in new_num_comm:
    for j in new_num_comm[i]:
        for k in present:
            if i==k[0] and k[1]==j:
                new_num_comm[i][j].append(new_arr[i][k[2]])
```

```python
for i in new_num_comm:
    for j in new_num_comm[i]:
        plt.figure()
        plt.title('Precision for seed node '+str(i))
        plt.xlabel('k')
        plt.ylabel('Fraction of points in community '+str(j))
        
        frac = [i1/len(node_conductance[i]) for i1 in range(1,len(new_num_comm[i][j])+1)]
        plt.plot(range(len(new_num_comm[i][j])),frac)
        
```

### Second Interpretation of Precision - In Order

```python
for i in new_num_comm:
    for j in new_num_comm[i]:
        n_arr = []
        for k in present:
            if k[0]==i and k[1]==j:
                n_arr.append(k[2])
        tmp = (len(new_arr[i]))
        x = range(len(new_arr[i]))
        y = []
        j1=0
        val=0
        for i1 in range(len(new_arr[i])):
            if j1 in n_arr:
                val+=1
                y.append(val/(j1+1))
            else:
                if val-1!=0:
                    val-=1
                    y.append(val/(j1+1))
                else:
                    y.append(val/(j1+1))
            j1+=1
        plt.figure()
        plt.xlabel('K')
        plt.ylabel('Precision')
        plt.title('Seed Node - '+str(i) + ' community '+str(j))
        plt.plot(x,y)
```
